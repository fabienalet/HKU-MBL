In this notebook, we consider the so-called PXP model
$$
    H_\text{PXP} = \sum_{j=1}^L P \sigma^x_j P
$$
where $P$ is the projector on spin configurations without ajdacent up spins:
$$
    P = \prod_j \left[ 1 - \frac{1}{4}(1+\sigma_j^z)(1+\sigma_{j+1}^z) \right].
$$
We will also consider some generalizations of the model.

This notebook is mainly based on:
* [Nat. Phys. **14**, 745–749 (2018)](https://www.nature.com/articles/s41567-018-0137-5)
* [Phys. Rev. Lett. **122**, 220603 (2019)](https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.122.220603).

# Building the constrained basis
$P$ constrains the dynamics to a sub-Hilbert space of the chain of spins.
This constraint is numerically beneficial: by reducing the Hilbert space dimension from $2^L$ to $\sim \phi^L$ (where $\phi = (1+\sqrt{5})/2$ is the golden ratio), it lets us simulate larger systems.

As usual, we work in the 'computational basis' whose vectors are the simultaneous eigenstates of the $\sigma_j^z$ Pauli operators.
Let us first write a simple class storing such configurations, and raising an exception if the current configuration does not satisfy the constraint.

```python
class Conf():
    """
    Stores a valid configuration of the constrained Hilbert space
    """
    def __init__(self, qubits: str):
        if '11' in qubits:
            raise KeyError(f'Invalid configuration {qubits}')
        self._qubits = qubits

    def __add__(self, conf):
        return Conf(self._qubits + conf._qubits)

    def __repr__(self):
        return self._qubits

    def __hash__(self):
        """
        Associate a unique integer to each configuration (useful for building operators)
        """
        return int(self._qubits, 2)
```

We decompose the set of configurations of the system of a given length into 4 ''sectors'' $S_{s_i,s_f}$, defined as the set of configurations whose first/last spin is fixed to $s_i$/$s_f$.
Writing recursion relations for these sectors is straightforward (for an illustration see Fig. 5. of [SciPost Phys. **6**, 033 (2019)](https://scipost.org/10.21468/SciPostPhys.6.3.033)).
The code below implements them.

```python
zero = Conf('0')
one = Conf('1')
def next_sectors(S00, S01, S10, S11):
    """
    Return sectors obtained by adding one spin to the right of the input sectors
    """
    return [C + zero for C in S00 + S01], [C + one for C in S00], [C + zero for C in S10 + S11], [C + one for C in S10]

@lru_cache(maxsize=128)
def sectors(L: int):
    """
    Return sectors of the system of L spins
    """
    if L < 2:
        raise RuntimeError(f'Sectors are ill-defined for a chain of size L = {L}.')
    S00 = [zero + zero]
    S01 = [zero + one]
    S10 = [one + zero]
    S11 = []
    for _ in range(L-2):
        S00, S01, S10, S11 = next_sectors(S00, S01, S10, S11)
    return S00, S01, S10, S11

def basis_vectors(L: int):
    """
    Return a list of basis vectors of constrained system of L spins, using open boundary conditions
    """
    if L == 0:
        return []
    if L == 1:
        return [zero, one]
    return np.concatenate(sectors(L))

def size11(L: int):
    """
    Return the size of the 11 sector
    """
    if L == 0 or L == 1:
        return 0
    return len(sectors(L)[3])
```

We can check that the Hilbert space sizes are consecutive Fibonacci numbers.

```python
for L in range(1,5):
    print(len(basis_vectors(L)))
```

Let us now wrap these methods in a `Basis` class.

```python
class Basis:
    """
    Store all configurations on the constrained chain of L sites
    """
    def __init__(self, L: int):
        # number of qubits
        self._L = L
        # list of confs
        self._confs = basis_vectors(L)
        # Hilbert space size for OBC
        self._Nobc = len(self._confs)
        # Hilbert space size for PBC
        self._Npbc = self._Nobc - size11(L)
        # index of the configurations
        self._InverseMap = {hash(conf):i for i, conf in enumerate(self._confs)}

    @property
    def L(self):
        return self._L

    @property
    def Nobc(self):
        return self._Nobc

    @property
    def Npbc(self):
        return self._Npbc

    @property
    def confs(self):
        return self._confs

    def index(self, conf):
        """
        Return index of the conf
        """
        return self._InverseMap[hash(conf)]
```

# Building the Hamiltonian

We are now ready to build operators on the constrained subspace. 
We will make use of the `Operator` class in `lib/Constrained` to store operators in a sparse format.

Let us first build the local $\sigma^x$ and $\sigma^z$ operators we shall need later to construct the PXP Hamiltonian.

```python
data = [1.,1.]
zero = Conf('0')
one = Conf('1')
row_conf = [zero, one]
col_conf = [one, zero]
X = ct.Operator((data, (row_conf, col_conf)), Basis(1))

data = [-1.,1.]
row_conf = [zero, one]
Z = ct.Operator((data, (row_conf, row_conf)), Basis(1))
```

The `kron` method implements the Kronecker (or outer) product on the constrained Hilbert space. Using it, we can construct operators acting on the chain of $L$ site from the previously defined local operators, and from them the PXP Hamitonian.

```python
def identity(L: int):
    basis = Basis(L)
    row_conf = basis.confs
    data = [1.]*basis.Nobc
    return ct.Operator((data, (row_conf, row_conf)), basis)

def Xop(basis: Basis, j: int):
    """
    X operator acting on site j on a chain of size L
    """
    if j >= basis.L:
        raise KeyError(f'Index i={j} >= L={basis.L}')
    # left part of the operator
    op = ct.kron(identity(j), X)
    # right part of the operator
    return ct.kron(op, identity(basis.L-j-1))

def PXP(basis):
    """
    Return the PXP Hamiltonian, with open or periodic boundary conditions
    """
    H0 = Xop(basis, 0).tocoo()
    for i in range(1, basis.L):
        H0 += Xop(basis, i).tocoo()
    return H0.todense()
```

# Scars

The PXP Hamiltonian hosts eigenstates with anomalously high overlap with simple product states such as the Néel or $|\mathbb{Z}_2\rangle$ state. Such eigenstates, although at finite energy density, have low entanglement, and are close to being **many-body scars**, eigenstates violating the eigenstate thermalization hypothesis.

Below, we identify such eigenstates by plotting the overlap with the $|\mathbb{Z}_2\rangle$ state.

```python
def get_Z2_index(basis):
    Z2 = ''
    for i in range(basis.L//2):
        Z2 += '01'
    return basis.index(ct.Conf(Z2))

def get_eigenrpr(basis, P):
    """
    Return the Z2 state eigenbasis decomposition
    """
    Z2_index = get_Z2_index(basis)
    return np.array(P[Z2_index])[0]
```

```python
L = 14
basis = Basis(L)
# construct PXP Hamiltonian
H = PXP(basis)
# diagonalize the PXP Hamiltonian with PBC
# Rk: to obtain the PBC Hamiltonian from the OBC, we project out the 11 sector
E, P = np.linalg.eigh(H[:basis.Npbc,:basis.Npbc])
```

```python
fig, ax = plt.subplots()
# compute overlap between Z2 state and the eigenstates of the PXP Hamitonian
eigenrpr = get_eigenrpr(basis, P)
overlaps = np.abs(eigenrpr)**2.
ax.plot(E, overlaps, 'o')
ax.set_xlabel('$E$')
ax.set_ylabel(r'$|\langle \mathbb{Z}_2 | E \rangle|^2$')
ax.set_yscale('log')
ax.set_ylim(1e-12, 10)
plt.show()
```

Because the $|\mathbb{Z}_2\rangle$ state mainly lives in a low-dimensional scar manifold, its dynamics shows long-lived Rabbi-like oscillations. Below, we observe such oscillations by computing the time evolution of the fidely from the full diagonalization of the PXP Hamiltonian.

```python
def fidelities(basis, E, P, times):
    """
    Compute fidelities for a given timerange
    """
    eigenrpr = get_eigenrpr(basis, P)
    Z2_index = get_Z2_index(basis)
    fidelities = []
    for t in times:
        state = P @ (np.exp(-1j*t*E) * eigenrpr)
        fidelity = np.abs(state.T[Z2_index])**2
        fidelities.append(fidelity[0,0])
    return fidelities
```

```python
times = np.linspace(0, 30, 500)
fid = fidelities(basis, E, P, times)
```

```python
fig, ax = plt.subplots()
ax.plot(times, fid)
ax.set_xlabel('$t$')
ax.set_ylabel('$g(t)$')
plt.show()
```

Fidelity revivals are not perfect, signalling leakage out of the scar manifold.
Moreover, the fidelity at a given time drops with increasing system size, a further hint at the imperfection of the many-body scars of the PXP model.

```python
L = 16
basis = Basis(L)
# construct PXP Hamiltonian
H = PXP(basis)
# diagonalize the PXP Hamiltonian with PBC
# Rk: to obtain the PBC Hamiltonian from the OBC, we project out the 11 sector
E, P = np.linalg.eigh(H[:basis.Npbc,:basis.Npbc])

times = np.linspace(0, 30, 500)
fid16 = fidelities(basis, E, P, times)
```

```python
fig, ax = plt.subplots()
ax.plot(times, fid,label='L=14')
ax.plot(times, fid16,label='L=16')
ax.set_xlabel('$t$')
ax.set_ylabel('$g(t)$')
plt.legend()
plt.show()
```

# ZPXP models
We can bring the model closer to being perfectly scarred, by adding new longer-range ZPXP couplings.
Following [Nat. Phys. **14**, 745–749 (2018)](https://www.nature.com/articles/s41567-018-0137-5), we consider two generalizations of the model, both involving longer range operators $\text{PXPZ}_j(d) = P\sigma_j^xP(\sigma_{j-d}^z + \sigma_{j+d}^z)$.
1. A model $H_2 = H_\text{PXP} + \delta H_2$ obtained by adding next-to-nearest neighbors couplings:
$$
    \delta H_2 = -h_2 \sum_j \text{PXPZ}_j(2),
$$
2. A model $H_d = H_\text{PXP} + \delta H_R$ obtained by adding couplings at all distances:
$$
    \delta H_R = - \sum_{d=2}^{R=L//2} h_d \sum_j \text{PXPZ}_j(d),
$$
for some $h_d$ whose expression is given below.

Let us write functions returning the various operators needed to build the above models.
For that purpose, we make an extensive use of the `kron` routine to compute outer products of local operators.

```python
def two_ops(basis, i: int, shift: int, Op1, Op2):
    """
    Return the operator made out of the product of Op1 at site i and Op2 at site i+shift.
    """
    Op = ct.kron(identity(i), Op1)
    Op = ct.kron(Op, identity(shift-1))
    Op = ct.kron(ct.kron(Op, Op2), identity(basis.L-i-shift-1))
    return Op

def PXPZmat(basis, j, d):
    """
    Return the matrix corresponding to operator PXPZ_i(shift)
    """
    return two_ops(basis, j, d, X, Z).tocoo() + two_ops(basis, j, d, Z, X).tocoo()

def HZX(basis, d, pbc: bool):
    """
    Return the Hamiltonian sum_j(PXPZ_j(d))
    """
    HXZ = PXPZmat(basis, 0, d)
    for i in range(1, L-d):
        HXZ += PXPZmat(basis, i, d)
    if pbc:
        for s in range(d):
            HXZ += PXPZmat(basis, s, L-d)
    return HXZ
```

Now, we can build the 3 Hamiltonians: $H_\text{PXP}$, $H_2$ and $H_R$.

```python
h0 = -0.051
phi = 0.5*(1. + np.sqrt(5))
def hAnsatz(d: int):
    return h0*(phi**(d-1.) - phi**(-d+1.))**(-2.)
```

```python
L = 14
pbc = True
basis = Basis(L)
HPXP = PXP(basis)
# range 2 ZPXP terms
h2 = 0.053
H2 = HPXP - h2*HZX(basis, 2, pbc)
# all range ZPXP terms
HR = HPXP
for d in range(2, L//2):
    HR += hAnsatz(d) * HZX(basis, d, pbc)
```

```python
Es, Ps = [], []
Hams = [HPXP, H2, HR]
for Ham in Hams:
    tmp_E, tmp_P = np.linalg.eigh(Ham[:basis.Npbc,:basis.Npbc])
    Es.append(tmp_E)
    Ps.append(tmp_P)
```

We see that adding correction terms improves the overlap of the scarred eigenstates with the Neel state.

```python
fig, ax = plt.subplots()
labels = ['PXP', 'ZPXP$_2$', 'ZPXP$_R$']
markers = ['o', '^', '>']
for E, P, label, marker in zip(Es, Ps, labels, markers):
    eigenrpr = get_eigenrpr(basis, P)
    overlaps = np.abs(eigenrpr)**2.
    ax.plot(E, overlaps, lw=0, marker=marker, label=label)
ax.set_xlabel('$E$')
ax.set_ylabel(r'$|\langle \mathbb{Z}_2 | E \rangle|^2$')
ax.set_yscale('log')
ax.set_ylim(1e-12, 10)
plt.legend()
plt.show()
```

As expected, this improved overlap corresponds to better revivals of the fidelity when the system is prepared in its Néel state. The $H_R$ Hamiltonian in particular features almost perfect revivals for the system size and the time scales probed here.

```python
times = np.linspace(0, 30, 500)
fids = []
for E, P in zip(Es, Ps):
    tmp_fid = fidelities(basis, E, P, times)
    fids.append(tmp_fid)
```

```python
fig, ax = plt.subplots()
labels = ['PXP', 'ZPXP$_2$', 'ZPXP$_R$']
for fid, label in zip(fids, labels):
    ax.plot(times, fid, label=label)
ax.set_xlabel('$t$')
ax.set_ylabel('$g(t)$')
plt.legend()
plt.show()
```
